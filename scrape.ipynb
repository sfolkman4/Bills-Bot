```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
```

```python
url = 'https://www.congress.gov/bill/118th-congress/house-bill/6363/text?s=3&r=1'
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get(url)
```

```python
# Locate the element with the ID 'billTextContainer'
bill_text_element = driver.find_element_by_id('billTextContainer')

# Extract the text content from the element
bill_text = bill_text_element.text

# Print or use the extracted text
print(bill_text)
```

```python
# Close the browser
driver.quit()
```
